```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

### Gathering and Assessing Data

```python
# load in the dataset into a pandas dataframe, print statistics
trip = pd.read_csv('201902-fordgobike-tripdata.csv')
trip.head(10)
```

```python
# high-level overview of data shape and composition
print(trip.shape)
print(trip.dtypes)
```

```python
trip.info()
```

```python
# change the data type for start and end time to be datetime
trip['start_time'] = pd.to_datetime(trip['start_time'])
trip['end_time'] = pd.to_datetime(trip['end_time'])
```

```python
# change the data type for bike_share_for_all_trip to be bool
trip['bike_share_for_all_trip'] = (trip['bike_share_for_all_trip'] == 'Yes')
```

```python
# check the data types after the changes
trip.info()
```

```python
# Add a new column calculating users age from 'member_birth_year'
trip['member_age'] = 2021 - trip['member_birth_year']

# filter out any outlier ages
trip = trip.query('member_age <= 70')

# casting 'member_birth_year' and 'member_age' to integer instead of float type
trip['member_birth_year'] = trip['member_birth_year'].astype('int')
trip['member_age'] = trip['member_age'].astype('int')
```

```python
# Add a new columns calculating start_hour_of_day, start_day and start_month
trip['start_hour_of_day'] = trip['start_time'].dt.strftime('%H')
trip['start_day'] = trip['start_time'].dt.strftime('%A')
```

## Univariate Exploration

> I will start by looking at the distribution of the main column of interest which is: duration_sec

```python
binsize = 500
bins = np.arange(0, trip['duration_sec'].max() + binsize, binsize)

plt.figure(figsize=[8, 5])
plt.hist(data = trip, x = 'duration_sec', bins = bins)
plt.title('Distribution of Trip Durations')

# Labeling the axes
plt.xlabel('Trip Duration (sec)')
plt.ylabel('Count of Trips')

plt.axis([-500, 6000, 0, 90000])

# Dsiplay the plot
plt.show()
```

We can not see clearly which the data looks highly skewed  with a very long tail of data points, so we will Scale the x-axis to log-type scales for more clear vision

```python
# let us take the log scale for more clear vision

bins_log = 10 ** np.arange(2.4, np.log10(trip['duration_sec'].max()) + 0.05, 0.05)

plt.figure(figsize=[8, 5])
plt.hist(data = trip, x = 'duration_sec', bins = bins_log)

# The argument in the xscale() represents the axis scale type to apply.
plt.xscale('log')

plt.title('Distribution of Trip Durations')

# Labeling the axes
plt.xlabel('Trip Duration (sec)')
plt.ylabel('count of Trips')

# show the scale in x-axis will be 1K instead of 1000, and so on..
plt.xticks([500, 1e3, 2e3, 5e3, 1e4], [500, '1k', '2k', '5k', '10k'])

# set limits for y and x axis
plt.axis([200, 10000, 0, 14000])

# Dsiplay the plot
plt.show()
```

* Number of Trips increases starting from around 8000 to 12000 values, then starts to fall down around to 2000 values approximately.
* The duration of the trip is approximately centered on values which are less than 2000 seconds with peak around 500 seconds.

> In the beginning, I want to look at the duration_sec (which is the most strongest feature) correlations present between features such as: age, start station, end station, user type, and gender.



```python
# plotting Trip Duration and Age

plt.figure(figsize=[16,5])

plt.subplot(1, 2, 1)
plt.scatter(data = trip, x = 'member_age', y = 'duration_sec', alpha=1/4)
plt.title('Trip Duration and Age')
plt.axis([10, 90, 500, 7500])

# Labeling the axes
plt.xlabel('Age')
plt.ylabel('Duration in Seconds')

plt.subplot(1, 2, 2)
bins_y = np.arange(500, 8000, 1000)
bins_x = np.arange(10, 94, 10)

# Use cmin to set a minimum bound of counts 
# Use cmap to reverse the color map. 
plt.hist2d(data = trip, x = 'member_age', y = 'duration_sec', cmin = 0.5, cmap='viridis_r', bins = [bins_x, bins_y])
plt.colorbar(ticks=[5000, 10000 , 15000, 20000, 25000, 30000, 35000], label = 'Duration (Sec)')
plt.title('Trip Duration and Age')

# Labeling the axes
plt.xlabel('Age')
plt.ylabel('Trip Duration (sec)')

# Dsiplay the plot
plt.show()
```

By investigating the previous subplots, we can observe that most frequent Age of users is between 20 and 40.

## Multivariate Exploration

> The main thing I want to explore in this part of the analysis is:
> * How does the trip duration vary in weekdays between customers and subscribers?
> * how the three categorical measures of user type and gender into the relationship between trip duration and age?


```python
plt.figure(figsize = [10, 5])

sb.pointplot(data=trip, x='start_day', y='duration_minute', hue='user_type', dodge=0.5, linestyles="") #dodge: Amount to separate the points
plt.xlabel('Day of Week') 
plt.ylabel('Trip Duration (mins)')
```

* we can see that subscribers ride much quicker trips compared to customers on each day of the week. 
* Both of user types have an evident increase in trip duration on Saturdays and Sundays which are the days of weekends, especially for customers.

```python
plt.suptitle('Hourly Usage during Weekdays for Customers and Subscribers')
plt.figure(figsize = [15, 10])

# subplot 1
plt.subplot(2, 1, 1)
customers = trip.query('user_type == "Customer"')
ct_counts = customers.groupby(['start_day', 'start_hour_of_day']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index='start_day', columns='start_hour_of_day', values='count')

sb.heatmap(ct_counts, cmap='viridis_r')

plt.title('Customer', loc='right')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')

# subplot 2 
plt.subplot(2, 1, 2)
sub = trip.query('user_type == "Subscriber"')
sub_counts = sub.groupby(['start_day', 'start_hour_of_day']).size()
sub_counts = sub_counts.reset_index(name='count')
sub_counts = sub_counts.pivot(index='start_day', columns='start_hour_of_day', values='count')

sb.heatmap(sub_counts, cmap='viridis_r')

plt.title('Subscriber', loc='right')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
```

Subscribers use the system profusely on work days (Monday to Friday) while customers use it a lot on weekends, especially in the afternoon. Many trips concentrated around 8 and 8 AM and 17 and 18 PM on work days for subscribers when used for work transport, yet customers tend to use more in the late afternoon approx. 17 PM Monday to Friday.

```python
t = sb.FacetGrid(data = trip, col = 'user_type', col_wrap = 2, height = 5,
                 xlim = [10, 80], ylim = [0, 250], margin_titles = True)

t.map(plt.scatter, 'member_age', 'duration_minute', alpha= 1/2)

t.set_xlabels('Age')
t.set_ylabels('Trip Duration (mins)')
t.fig.suptitle('Trip Duration, User type, and Age', va = 'center')
plt.show()
```

For customers with age between around 20 to 40, the duration of their trips is around 100 mins, then the duration drops during increasing in the age of customers.
