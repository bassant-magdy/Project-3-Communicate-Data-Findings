```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

### Gathering and Assessing Data

```python
# load in the dataset into a pandas dataframe, print statistics
trip = pd.read_csv('201902-fordgobike-tripdata.csv')
trip.head(10)
```

```python
# high-level overview of data shape and composition
print(trip.shape)
print(trip.dtypes)
```

```python
trip.info()
```

```python
# change the data type for start and end time to be datetime
trip['start_time'] = pd.to_datetime(trip['start_time'])
trip['end_time'] = pd.to_datetime(trip['end_time'])
```

```python
# change the data type for bike_share_for_all_trip to be bool
trip['bike_share_for_all_trip'] = (trip['bike_share_for_all_trip'] == 'Yes')
```

```python
# check the data types after the changes
trip.info()
```

```python
# Add a new column calculating users age from 'member_birth_year'
trip['member_age'] = 2021 - trip['member_birth_year']

# filter out any outlier ages
trip = trip.query('member_age <= 70')

# casting 'member_birth_year' and 'member_age' to integer instead of float type
trip['member_birth_year'] = trip['member_birth_year'].astype('int')
trip['member_age'] = trip['member_age'].astype('int')
```

```python
# Add a new columns calculating start_hour_of_day, start_day and start_month
trip['start_hour_of_day'] = trip['start_time'].dt.strftime('%H')
trip['start_day'] = trip['start_time'].dt.strftime('%A')
```

##  Distribution of Trip Durations

Trip duration is in second and in the dataset taking on a very large range of values so We can not see clearly which the data looks highly skewed with a very long tail of data points, so we will Scale the x-axis to log-type scales for more clear vision. 
we will observe that Number of Trips increases starting from around 8000 to 12000 values, then starts to fall down around to 2000 values approximately.
The duration of the trip is approximately centered on values which are less than 2000 seconds with peak around 500 seconds.

```python
# let us take the log scale for more clear vision

bins_log = 10 ** np.arange(2.4, np.log10(trip['duration_sec'].max()) + 0.05, 0.05)

plt.figure(figsize=[8, 5])
plt.hist(data = trip, x = 'duration_sec', bins = bins_log)

# The argument in the xscale() represents the axis scale type to apply.
plt.xscale('log')

plt.title('Distribution of Trip Durations')

# Labeling the axes
plt.xlabel('Trip Duration (sec)')
plt.ylabel('count of Trips')

# show the scale in x-axis will be 1K instead of 1000, and so on..
plt.xticks([500, 1e3, 2e3, 5e3, 1e4], [500, '1k', '2k', '5k', '10k'])

# set limits for y and x axis
plt.axis([200, 10000, 0, 14000])

# Dsiplay the plot
plt.show()
```

##  Distribution of User Type

User Type has two categorical variables (Customer and Subscriber) and we will observe that The number of subscribers are much larger than the number of customers.

## Distribution of Hours over Days

The trip distribution over hours peaks during rush hours of work between 8AM and 9AM and between 17PM to 18PM.

## Distribution of Days over Week

The trip distribution over day of week, it is quite obvious that the most of trips happened on work days from Monday to Friday and many usage is probably for commuting and we can also see that The distribution decreases during holidays (the Weekends outside Egypt is during Saturday and Sunday).

```python
# Returns the types for 'days' with the categories and orderedness
days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daycat = pd.api.types.CategoricalDtype(ordered=True, categories = days)
trip['start_day'] = trip['start_day'].astype(daycat)

# Plotting
plt.figure(figsize=[8, 5])

base_color = sb.color_palette()[0]
sb.countplot(data=trip, x='start_day', color= base_color)
plt.title('Distribution of Days over Week')

# Labeling the axes
plt.xlabel('Trip Start Day of Week')
plt.ylabel('Count of Trips')
```

## Trip Duration vs. Age

By investigating the next subplots, we can observe that most frequent Age of users is between 20 and 40.

## Hourly Usage during Weekdays by Customers and Subscribers

The next plot shows obviously very different usage between the two type of users. Subscribers use the system profusely on work days (Monday to Friday) while customers use it a lot on weekends, especially in the afternoon. Many trips concentrated around 8 and 8 AM and 17 and 18 PM on work days for subscribers when used for work transport, yet customers tend to use more in the late afternoon approx. 17 PM Monday to Friday.

```python
# How does the hourly usage vary during weekdays for customers and subscribers?

plt.suptitle('Hourly Usage during Weekdays for Customers and Subscribers')
plt.figure(figsize = [15, 10])

# subplot 1
plt.subplot(2, 1, 1)
customers = trip.query('user_type == "Customer"')
ct_counts = customers.groupby(['start_day', 'start_hour_of_day']).size()
ct_counts = ct_counts.reset_index(name='count')
ct_counts = ct_counts.pivot(index='start_day', columns='start_hour_of_day', values='count')

sb.heatmap(ct_counts, cmap='viridis_r')

plt.title('Customer', loc='right')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')

# subplot 2 
plt.subplot(2, 1, 2)
sub = trip.query('user_type == "Subscriber"')
sub_counts = sub.groupby(['start_day', 'start_hour_of_day']).size()
sub_counts = sub_counts.reset_index(name='count')
sub_counts = sub_counts.pivot(index='start_day', columns='start_hour_of_day', values='count')

sb.heatmap(sub_counts, cmap='viridis_r')

plt.title('Subscriber', loc='right')
plt.xlabel('Hour of Day')
plt.ylabel('Day of Week')
```
